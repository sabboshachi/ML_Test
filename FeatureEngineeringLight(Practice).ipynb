```
sex_dummy=pd.get_dummies(data['sex'], drop_first= True)
sex_dummy
```

```
smoker_dummy=pd.get_dummies(data['smoker'], drop_first= True)
smoker_dummy
```

```
region_dummy=pd.get_dummies(data['region'], drop_first= True)
region_dummy
```

```
df=pd.concat([sex_dummy,region_dummy,smoker_dummy,data['age'],data['bmi'],data['children'],data['charges']], axis=1)
df
```
