```python
sex_dummy=pd.get_dummies(data['sex'], drop_first = True)
sex_dummy
```

```python
smoker_dummy=pd.get_dummies(data['smoker'], drop_first = True)
smoker_dummy
```

```python
region_dummy = pd.get_dummies(data["region"], drop_first = True)
region_dummy
```

```python
df = pd.concat([sex_dummy, smoker_dummy, region_dummy, data["age"], data["bmi"], data["children"],data["charges"]], axis = 1)
df
```

```python
df.info()
```

#Linear Regression
From here the part of normal calculation begins. As all the datas in the data table are numerical values, we can easily apply any model to find out the requried result.




```python
x = df.drop(['charges'], axis=1)
y = df['charges']
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.10)  
```

```python
from sklearn.linear_model import LinearRegression

linear = LinearRegression() 

linear.fit(x_train,y_train)
```
