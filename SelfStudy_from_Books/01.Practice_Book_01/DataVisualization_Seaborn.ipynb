```python
import seaborn as sns
```

```python
iris = sns.load_dataset('iris')
```

```python
iris
```

```python
correlation = iris.corr()
```

```python
correlation
```

**Standard Correlation Coefficient** সাধারণতঃ -1 থেকে 1 হয়। মানে **Linear Coefficient** হচ্ছে যখন একটা বাড়ে, তখন আরেকটা জিনিস সেভাবে বাড়ে বা কমে। 
যখন এটা 1 এর কাছাকাছি হয় তখন সেটাকে একটা **Strong Positive Correlation** হিসেবে ধরা হয়। এদিকে সেটা যদি -1 এর দিকে যায়, তখন সেটাকে আমরা **Strong Negative 
Correlation** বলি। আবার 0 এর কাছে হলে এটার কোন **Linear Correlation** নেই। petal_length এর সাথে petal_width এর সম্পর্ক 0.962865 মানে ১ এর কাছাকাছি। sepal_length এর সাথে petal_length এর সম্পর্ক 0.871754 হলে Strong Positive Correlation আছে বলে ধরা হয়।

```python
sns.heatmap(iris.corr(), cmap = "coolwarm", annot = True)
```

```python
sns.set(style="whitegrid", color_codes=True) 
```

```python
sns.lmplot(x="sepal_length", y="petal_length", hue="species", data = iris, palette="Set1", markers=["o", "s", "d"])
```

```python
sns.lmplot(x="sepal_length", y="petal_length", hue="species", data = iris, palette="Set1", fit_reg=False, markers=["o", "s", "d"])
```

```python
sns.pairplot(iris, size=5, vars=["sepal_length", "petal_length"], markers=["o", "s", "d"], hue= "species")
```

```python
sns.pairplot(iris, markers=["o", "s", "d"], hue="species")
```

```python
sns.pairplot(iris, kind="reg", hue="species", markers=["o", "s", "d"])
```

#Some more plots for better understanding...

```python
iris["ID"] = iris.index
iris["ratio"] = iris["sepal_length"]/iris["sepal_width"]
sns.lmplot(x="ID", y="ratio", data=iris, hue="species", markers=["o", "s", "d"], fit_reg=False, legend=True)
```
